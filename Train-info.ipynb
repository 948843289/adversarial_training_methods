```python
import numpy as np
import matplotlib.pyplot as plt

x = np.load('../dataset/imdb/nltk_xtrain.npy')
ul = np.load('../dataset/imdb/nltk_ultrain.npy')
len_x = np.asarray( [len(s) for s in x] ).astype('f')
len_ul = np.asarray( [len(s) for s in ul] ).astype('f')

threshold = 600 

print( 'mean: {:.1f}'.format(len_x.mean()) )
print( 'standard deviation: {:.1f}'.format(len_x.std()) )
print( 'max len: {}'.format(max(len_x)) )
print( 'min len: {}'.format(min(len_x)) )
print( 'percentage of sequences with length <= of {} is {:.3f}'.format(threshold, len([s for s in len_x if (s<=threshold)])/len(len_x)) )
hist = np.histogram(len_x, bins=int(max(len_x)))
fig = plt.figure()
a = fig.add_subplot(111)
a.plot( hist[0] )
a.set_xlabel('sequence length')
a.set_ylabel('number of examples')
a.axvline(x=threshold, color='r', linestyle='--', linewidth=0.7)
a.axvline(x=400, color='g', linestyle='--', linewidth=0.7)
#fig.savefig('img/train_info_both_plot.png')
plt.show()

print(); print('-----------------------------------------'); print()

print( 'mean: {:.1f}'.format(len_ul.mean()) )
print( 'standard deviation: {:.1f}'.format(len_ul.std()) )
print( 'max len: {}'.format(max(len_ul)) )
print( 'min len: {}'.format(min(len_ul)) )
print( 'percentage of sequences with length <= of {} is {:.3f}'.format(threshold, len([s for s in len_ul if (s<=threshold)])/len(len_ul)) )
hist = np.histogram(len_x, bins=int(max(len_ul)))
plt.plot( hist[0] )
plt.show()
```
